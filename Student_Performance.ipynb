```python
import pandas as pd
# Creating  a sample dataset
data = {
    "Name": ["Aman", "Bhavya", "Chirag", "Divya", "Esha", "Farhan", "Gauri", "Harshit", "Isha", "Jay"],
    "Gender": ["M", "F", "M", "F", "F", "M", "F", "M", "F", "M"],
    "Math": [88, 76, 90, 65, 70, 85, 60, 92, 74, 80],
    "Science": [75, 85, 78, 70, 82, 88, 65, 90, 80, 76],
    "English": [80, 78, 75, 72, 84, 70, 68, 85, 90, 73],
    "Attendance": [90, 85, 95, 80, 88, 92, 78, 96, 89, 87]
}
df = pd.DataFrame(data)
df.to_csv("student_data.csv", index=True)
df
```

```python
import pandas as pd
df=pd.read_csv("student_data.csv")
df.head(5)
```

```python
#Average marks in each subject
avg_math=df["Math"].mean()
avg_science=df["Science"].mean()
avg_english=df["English"].mean()

print("Average Maths marks :",avg_math)
print("Average Science marks :",avg_science)
print("Average English marks :",avg_english)
```

```python
# Find the topper in each subject
top_science=df[df["Science"]==df["Science"].max()]
print("Science Topper")
print(top_science[["Name","Science"]])

top_math=df[df["Math"]==df["Math"].max()]
print("\nMath Topper")
print(top_math[["Name","Math"]])

top_english=df[df["English"]==df["English"].max()]
print("\n English Topper")
print(top_english[["Name","English"]])
```

```python
df["Total"]=df["Math"]+df["Science"]+df["English"]
df["Percentage"]=df["Total"]/3
df
```

```python
#visualizing data with matplotlib
import matplotlib.pyplot as plt
subjects=["Math","Science","English"]
averages=[avg_math,avg_science,avg_english]

#Bar Chart
plt.figure(figsize=(6,4))
plt.bar(subjects,averages,color='skyblue')
plt.title("Average marks in each subject")
plt.xlabel("Subjects")
plt.ylabel("Average Marks")
plt.ylim(0,100)
plt.show()
          
```

```python
#Pie Chart for gender distribution
gender_counts=df['Gender'].value_counts()
#Create pie chart
plt.figure(figsize=(5,5))
plt.pie(gender_counts,labels=gender_counts.index,autopct='%1.1f%%',colors=["lightcoral","lightskyblue"])
plt.title("Gender distribution in class")
plt.show()
```

```python
# Attendance vs Percentage
import matplotlib.pyplot as plt

df_sorted = df.sort_values(by="Attendance")

plt.figure(figsize=(8, 4))
plt.plot(df_sorted["Attendance"], df_sorted["Percentage"], marker='o', linestyle='-', color='green')
plt.title("Attendance vs Percentage")
plt.xlabel("Attendance (%)")
plt.ylabel("Percentage")
plt.grid(True)
plt.show()
```
